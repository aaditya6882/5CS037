# Task 1 - Sum of Nested Lists:
Scenario: You have a list that contains numbers and other lists of numbers (nested lists).
You want to find the total sum of all the numbers in this structure.
Task:

*   Write a recursive function sum_nested_list(nested_list) that:


1. Takes a nested list (a list that can contain numbers or other lists of numbers) as
input.
2. Sums all numbers at every depth level of the list, regardless of how deeply nested
the numbers are.
• Test the function with a sample nested list, such as
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]].
The result should be the total sum of all the numbers.

```python
def sum_nested_list(nested_list):
  '''
  This function will sum the values of the nested lists
  param:
    nested_list(list):nested list of numbers
  return:
    total_sum (int):total sum of all the numbers in the nested list
  '''
  total_sum = 0
  for item in nested_list:
    if isinstance(item, list):
      total_sum += sum_nested_list(item)
    else:
      total_sum+=item
  return total_sum
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
sum_nested_list(nested_list)
```

# Task 2 - Generate All Permutations of a String:
Scenario: Given a string, generate all possible permutations of its characters. This is useful
for understanding backtracking and recursive depth-first search.
### Task:
• Write a recursive function generate_permutations(s) that:


     Takes a string s as input and returns a list of all unique permutations.



• Test with strings like ”abc” and ”aab”.

    print(generate_permutations("abc"))
### Should return [’abc’, ’acb’, ’bac’, ’bca’, ’cab’, ’cba’]

```python
def generate_permutations(s):
  '''
  this function will generate all the permutations of the string
  param:
    s(str):string
  return:
    permutations_list(list):list of all the permutations
  '''
  if s == "":
    return [""]
  permutations_list=[];
  for char in s:
    generate_permutations_list=generate_permutations(s.replace(char, "", 1))
    for permutation in generate_permutations_list:
      permutations_list.append(char + permutation)
  return permutations_list
print(generate_permutations("abc"))
```

  *   directory is a dictionary where keys represent file names (with values as sizes in KB) or directory names (with values as another dictionary representing a subdi-

```python
def calculate_directory_size(directory):
  '''
  this function will calculate the total size of the directory
  param:
    directory(dict):dictionary of the directory
  return
    total_size(int):total size of the directory
  '''
  total_size = 0;
  for keys,values in directory.items():
    if isinstance(values,dict):
      total_size+=calculate_directory_size(values)
    else:
      total_size+=values
  return total_size
directory_structure = {
"file1.txt": 200,
"file2.txt": 300,
"subdir1": {
"file3.txt": 400,
"file4.txt": 100
},
"subdir2": {
"subsubdir1": {
"file5.txt": 250
},
"file6.txt": 150
}
}
print(calculate_directory_size(directory_structure));
```

# Task 1 - Coin Change Problem:
Scenario: Given a set of coin denominations and a target amount, find the minimum number
of coins needed to make the amount. If it’s not possible, return - 1.
### Task:


1.   Write a function min_coins(coins, amount) that:


*   Uses DP to calculate the minimum number of coins needed to make up the
amount.



2.   Test with coins = [1, 2, 5] and amount = 11. The result should be 3 (using coins [5, 5, 1]).



```python
def min_conis(coins, amount):
  '''
  this function will find the minimum number of coins needed to make the amount
  param:
    coins(list):list of coins
    amount(int):amount to make
  return:
    dp[amount](int):minimum number of coins needed to make the amount
  '''
  dp = [float('inf')] * (amount + 1)
  dp[0] = 0
  for coin in coins:
    for i in range(coin, amount + 1):
      dp[i] = min(dp[i], dp[i - coin] + 1)
  return dp[amount] if dp[amount] != float('inf') else -1
min_conis([1,2,5],11)
```

# Task 2 - Longest Common Subsequence (LCS):
Scenario: Given two strings, find the length of their longest common subsequence (LCS).
This is useful in text comparison.
## Task:
1. Write a function longest_common_subsequence(s1, s2) that:


*  Uses DP to find the length of the LCS of two strings s1 and s2.


2. Test with strings like "abcde" and "ace"; the LCS length should be 3 ("ace").

```python
def longest_common_subsequences(s1,s2):
  '''
  this function will find the length of the longest common subsequence of two strings
  param:
    s1(String):group of letters
    s2 (String): group of letters
  return:
    dp[m][n](int):length of longest common subsequence
  '''
  m=len(s1)
  n=len(s2)
  dp=[[0]*(n+1) for _ in range (m+1)]
  for i in range (1,m+1):
    for j in range (1,n+1):
      if (s1[i-1]==s2[j-1]):
        dp[i][j]=dp[i-1][j-1]+1
      else:
        dp[i][j]=max(dp[i-1][j],dp[i][j-1])
  return dp[m][n]
print(longest_common_subsequences("abcde","ace"))
```

# Task 3 - 0/1 Knapsack Problem:
Scenario: You have a list of items, each with a weight and a value. Given a weight capacity,
maximize the total value of items you can carry without exceeding the weight capacity.
## Task:
1. Write a function knapsack(weights, values, capacity) that:

*   Uses DP to determine the maximum value that can be achieved within the given
weight capacity.

2. Test with weights [1, 3, 4, 5], values [1, 4, 5, 7], and capacity 7. The re-
sult should be 9.

```python
def knapsack(weights,values,capacity):
  '''
  this function will find the maximum value that can be achieved within the given weight capacity
  param:
    weights(list):list of weights
    values(list):list of values
    capacity(int):weight capacity
  return:
    dp[n][capacity](int):maximum value that can be achieved
  '''
  n=len(values)
  dp=[[0 for _ in range (capacity+1)] for _ in range (n+1)]
  for i in range (1,n+1):
    for w in range (1,capacity+1):
      if weights[i-1]<=w:
        dp[i][w]=max(dp[i-1][w],values[i-1]+dp[i-1][w-weights[i-1]])
      else:
        dp[i][w]=dp[i-1][w]
  return dp[n][capacity]
print(knapsack([1,3,4,5],[1,4,5,7],7))
```
